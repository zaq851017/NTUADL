```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
list_p=[1.2,8.8,7.3,12.5,15.0,14.3,20.4,14.6,22.1,20.5,21.6,24.3,21.9,20.4,21.0,22.6,26.8,23.3,28.8,27.9,24.1,22.6,21.4,24.5,27.4]
```

```python
print(len(list_p))
```

```python
epoch = []
acc_p=[]
for i in range (20):
    epoch.append(40*i+160)
for i in range(20):
    sum_t = 0.0
    for j in range(i,i+5):
        sum_t += list_p[j]
    acc_p.append(sum_t/5)
```

```python
plt.plot(epoch,acc_p,'s-',color = 'r', label="pg")
plt.title("dqn learning curves of rewards with MA(n=5)")
plt.xlabel("episodes")
plt.ylabel("reward")
plt.show()
```
